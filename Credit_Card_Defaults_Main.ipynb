#Logistic Regression Prep

```python
df.head(2)
```

```python
#Prepare x and y
y = df["default payment next month"]
x = df.drop(columns=["ID", "default payment next month"])
x = sm.add_constant(x)
x.head(3)
```

```python
#Training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    random_state = 1502)
```

#Logistic Regression

```python
#Logistic Regression
model = sm.Logit(y_train, x_train).fit()
print(model.summary())
```

```python
#Create the function to read the output of Logistic regression
def logistic_reader(coefficient):
  probability = round((np.exp(coefficient)-1) * 100, 2)
  if probability > 0:
    print("The likelihood increases by", probability,"%")
  elif probability == 0:
    print("No impact")
  else:
    print("The likelihood decreases by", probability,"%")
```

#Assess the model

```python
#predictions
predictions = model.predict(x_test)
predictions = np.where(predictions > 0.5, 1,0)
predictions[:10]
```

```python
#Check if dataset is balance
y_test.mean()
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, predictions))
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
(tn, fp, fn, tp)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
#Accuracy KPI
accuracy = (tp + tn) / (tn + fp + fn + tp)
print("Accuracy:", round(accuracy * 100,2),"%")
```

```python
#F1 Score KPI
f1 = 2 * tp / (2 * tp + fn + fp)
print("F1 Score:", round(f1 * 100,2),"%")
```

```python
#Specificity KPI
specificity = tn / (tn + fp)
print("Specificity:", round(specificity * 100,2),"%")
```

```python
#Sensitivity KPI
sensitivity = tp / (tp + fn)
print("Sensitiivity:", round(sensitivity * 100,2),"%")
```
